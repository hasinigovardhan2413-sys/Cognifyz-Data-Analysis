* # Level 1 - Task 1: Top Cuisines
 ## OBJECTIVE:
The objective of this task is to analyze the restaurant dataset and identify the top three most common cuisines served by restaurants. This analysis also calculates the percentage of restaurants offering each of these cuisines and visualizes the results using a bar chart to understand customer performances and cuisine popularity. *




```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Dataset .csv')
```

```python
df.columns
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
top3 =  df['Cuisines'].value_counts().head(3)
```

```python
top3.plot(kind= 'bar')
plt.show()
```

```python
percentage = (top3 / len(df)) * 100
print(percentage)
```

** ## OBSERVATION:**

* - The top three most common cuisines were identified from the dataset.
 - Their percentages were calculated to understand how frequently they appear.
 - The bar chart clearly shows the popularity of each cuisine.
 - These cuisines represent the most common food preferences among the restaurants in the dataset. *
